```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import eurostat
```

```python
def get_deaths_per_year(country:str) -> pd.DataFrame:
    url = 'https://api.dane.gov.pl/1.4/resources/35335,liczba-zgonow-zarejestrowanych-w-rejestrze-stanu-cywilnego-w-okresie-od-1-wrzesnia-2015-r-dane-tygodniowe/data'
    df = eurostat.get_data_df('demo_r_mwk_ts')
    # T - all sexes
    df = df[(df['geo\\time'] == 'PL') & (df['sex'] == 'T')]
    
    # column should stand for a year
    sorted_columns = [x for x in df.columns.tolist() if x.startswith('2')]
    sorted_columns.sort()
    
    df = df[sorted_columns]
    
    df = df.T
        
    return df

def plot_data(df: pd.DataFrame):
    plt.plot(df)
    plt.xticks([])
```

```python
df = get_deaths_per_year('PL')
plot_data(df)
```
